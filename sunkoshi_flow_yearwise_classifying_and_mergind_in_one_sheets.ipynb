```python
df=pd.read_excel('/content/Final_DischargeData_Likhu (1).xlsx', sheet_name='Discharge_Data(all)', skiprows=[0])
```

```python
df['Date'] = pd.to_datetime(df['Date '])
```

```python
df['date'] = pd.to_datetime(df['Date']).dt.strftime('%Y-%b-%d')

# display the modified dataframe
print(df)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month

# iterate over each year and save monthly data into separate Excel files
for year in range(1964, 2021):
    year_data = df[df['year'] == year]  # filter data for the current year
    pivot_table = pd.pivot_table(year_data, values='Sunkoshi(652)', index='month', columns=year_data['Date'].dt.day, aggfunc='first').transpose()  # create pivot table for monthly data
    file_name = f'sunkoshi_monthly_data_{year}.xlsx'  # generate file name for the current year
    pivot_table.to_excel(file_name)
```

```python


# Define the starting year and d value
start_year = 1968
d = 1

# Create an empty list to store the dataframes
dfs = []

# Loop through each year, reading in the corresponding Excel file
for year in range(start_year, start_year + 52):
    filename = f"/content/sample_data/{year}.xlsx"
    df = pd.read_excel(filename)
    df['d'] = d
    df['year'] = year
    dfs.append(df)
    d += 1

# Concatenate the dataframes into a single dataframe
result = pd.concat(dfs)

# Print the first few rows of the result dataframe
print(result.head())
```
